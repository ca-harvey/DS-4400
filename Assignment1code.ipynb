### Problem 4.2 Code

#### Generate 3 Matrices

```python
matrix1 = np.random.randint(-10, 11, size = (3,3))
print(matrix1)
```

```python
matrix2 = np.random.randint(-10, 11, size = (3,3))
print(matrix2)
```

```python
matrix3 = np.random.randint(-10, 11, size = (3,3))
print(matrix3)
```

#### Compute Inverses

```python
matrix1Inv = np.linalg.inv(matrix1)
print(matrix1Inv)
```

```python
matrix2Inv = np.linalg.inv(matrix2)
print(matrix2Inv)
```

```python
matrix3Inv = np.linalg.inv(matrix3)
print(matrix3Inv)
```

#### Identity Matrices

```python
matrix1 @ matrix1Inv
```

```python
matrix2 @ matrix2Inv
```

```python
matrix3 @ matrix3Inv
```

## Problem 5 Code

#### Initial Setup of Data

```python
df = pd.read_csv("kc_house_data.csv")
```

```python
df.head()
```

```python
featuresOnly = df.drop(columns=["id", "date", "zipcode", "price"])
featuresOnly.head()
```

#### 5.1

```python
means = featuresOnly.mean()
means.sort_values()
```

```python
mins = featuresOnly.min()
mins.sort_values()
```

```python
maxes = featuresOnly.max()
maxes.sort_values()
```

```python
variance = featuresOnly.var()
variance.sort_values()
```

 #### 5.2 

```python
corr = featuresOnly.corrwith(df["price"])
corr.sort_values()
```

```python
corr_table = corr.to_frame(name="Correlation with Price")
corr_table.sort_values(by="Correlation with Price", ascending=True)
```
